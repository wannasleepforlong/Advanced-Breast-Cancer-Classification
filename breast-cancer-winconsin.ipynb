# 04. Data Correlation

```python
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
corr_matrix = data.corr()
```

```python
threshold = 0.7  # change as needed
target_corr = corr_matrix['diagnosis'].abs()  # <-- replace with actual target column name
important_features = target_corr[target_corr > threshold].index.tolist()
print("Selected Important Features:")
print(important_features)
filtered_corr = corr_matrix.loc[important_features, important_features]

plt.figure(figsize=(10,8))
sns.heatmap(filtered_corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Filtered Correlation Matrix (Important Features Only)")
plt.show()
```

# 05. Prepare Features & Target for Modeling

### 01. Features & Target

```python
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']
```

### 02. Standardize Features (Scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### 03. Train / Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# 06. Train Logistic Regression Model

```python
import time
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve

# ------------------------- MODELS -------------------------
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier

models = {
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(eval_metric='logloss', verbosity=0),
    "Neural Network (MLP)": MLPClassifier(max_iter=50)
}

# --- Color Palette ---
color_map = {
    "Random Forest": "royalblue",
    "Gradient Boosting": "seagreen",
    "XGBoost": "purple",
    "Neural Network (MLP)": "crimson"
}

results = {}
latency = {}
train_accuracy = {}

for name, model in models.items():
    print(f"\n================= {name} =================")

    # --- Training Time ---
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()

    latency[name] = end - start

    # Predictions
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Metrics
    train_accuracy[name] = accuracy_score(y_train, train_pred)
    results[name] = accuracy_score(y_test, test_pred)

    print(f"Training Time: {latency[name]:.4f}s")
    print(f"Training Accuracy: {train_accuracy[name]:.4f}")
    print(f"Test Accuracy: {results[name]:.4f}")

    # ---- Confusion Matrix ----
    cm = confusion_matrix(y_test, test_pred)
    plt.figure(figsize=(5,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap=sns.light_palette(color_map[name], as_cmap=True))
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

    # ---- ROC Curve ----
    roc_auc = roc_auc_score(y_test, y_prob)
    fpr, tpr, _ = roc_curve(y_test, y_prob)

    plt.figure(figsize=(5,3))
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})", color=color_map[name], linewidth=2)
    plt.plot([0,1], [0,1], 'k--')
    plt.title("ROC Curve")
    plt.legend()
    plt.show()

    print("*" * 50)


# --------- Training Loss Plots ---------
print("\n📌 Training Loss Curves")

# MLP Curve
mlp = models["Neural Network (MLP)"]
if hasattr(mlp, "loss_curve_"):
    plt.figure(figsize=(5,3))
    plt.plot(mlp.loss_curve_, color=color_map["Neural Network (MLP)"])
    plt.title("MLP Training Loss Curve")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.show()

# XGBoost Logloss
xgb = models["XGBoost"]
if hasattr(xgb, "evals_result_"):
    hist = xgb.evals_result_()
    if 'validation_0' in hist and 'logloss' in hist['validation_0']:
        plt.figure(figsize=(5,3))
        plt.plot(hist['validation_0']['logloss'], color=color_map["XGBoost"])
        plt.title("XGBoost Logloss")
        plt.xlabel("Epochs")
        plt.ylabel("Logloss")
        plt.show()


# --------- Summary Plot (Ascending Order) ---------
# Sort results by accuracy
sorted_items = sorted(results.items(), key=lambda x: x[1])
sorted_models = [x[0] for x in sorted_items]
sorted_values = [x[1] for x in sorted_items]

plt.figure(figsize=(7,4))
bars = plt.barh(sorted_models, sorted_values, color=[color_map[n] for n in sorted_models])
plt.title("Model Test Accuracy Comparison (Ascending)")
plt.xlabel("Accuracy")

# Add text labels
for i, v in enumerate(sorted_values):
    plt.text(v + 0.005, i, f"{v:.3f}", va='center')

plt.xlim(min(sorted_values) - 0.02, max(sorted_values) + 0.02)
plt.show()



# ---------- Final Summary Table ----------
print("\n==== Final Model Comparison ====")
for model in results:
    print(f"{model}: Test={results[model]:.4f}, Train={train_accuracy[model]:.4f}, Time={latency[model]:.4f}s")
```
